```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
x,y = datasets.make_blobs(n_samples=150,cluster_std=1.05,n_features=2,centers=2,random_state=2)
```

```python
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(14,7))
plt.scatter(x1[y == 0],x2[y == 0])
plt.scatter(x1[y == 1],x2[y == 1])
plt.show()
```

```python
para = {'w0':0,'w1':0,'w2':0}
```

```python
def step(z):
    return 0 if z <= 0 else 1
```

```python
def perceptron(lr,epochs):
    for j in range(epochs):
        for i in range(x.shape[0]):
            z = para['w1']*x1[i] + para['w2']*x2[i] + para['w0']
            y_hat = step(z)
            if (y_hat != y[i]):
                para['w1'] = para['w1'] + lr*(y[i]-z)*x1[i]
                para['w2'] = para['w2'] + lr*(y[i]-z)*x2[i]
                para['w0'] = para['w0'] + lr*(y[i]-z)    
```

```python
perceptron(0.5,100)
```

```python
m = -para['w1']/para['w2']
c = -para['w0']/para['w2']
```

```python
b = x1
```

```python
a = m*b + c
```

```python
plt.figure(figsize=(14,7))
plt.plot(b,a)
plt.scatter(x1[y == 0],x2[y == 0])
plt.scatter(x1[y == 1],x2[y == 1])
plt.show()
```

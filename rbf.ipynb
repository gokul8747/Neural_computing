```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_hidden = 10
n_outs = 1
```

```python
x = np.linspace(-1,1,100).reshape(-1,1)
y = np.sin(3*(x+0.5)**3-1) 
y += np.random.normal(0, 0.1, y.shape)  
```

```python
c_id = np.random.choice(x.shape[0],n_hidden)
```

```python
centers = x[c_id]
```

```python
def act(x,centers):
    z = []
    for i in centers:
        a = np.exp(-0.5*np.linalg.norm(x-i,axis=1)**2)
        z.append(a)
    return np.asarray(z)
```

```python
m = act(x,centers)
```

```python
w = y.T@np.linalg.pinv(m)
```

```python
y_pred = (w@m).T
```

```python
plt.figure(figsize=(14,7))
plt.plot(x,y)
plt.plot(x,y_pred)
plt.show()
```

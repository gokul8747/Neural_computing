```python
from sklearn import datasets
import numpy as np
```

```python
x,y = datasets.make_blobs(100,2,centers=2)
m,n = x.shape
x = x.reshape(n,m)
```

```python
input_size = 2
hidden_size = 3
output_size = 1
```

```python
wh = np.random.randn(hidden_size,input_size)
```

```python
wo = np.random.randn(output_size,hidden_size)
```

```python
bh = np.zeros((hidden_size,1))
```

```python
bo = np.zeros(1)
```

```python
weights = {"wh":wh,"wo":wo,"bh":bh,"bo":bo}
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def error(y,y_hat):
    return y-y_hat

def delt(y_hat):
    return y_hat*(1-y_hat)

def t_loss(y_hat,y):
    return np.mean(np.square(y-y_hat))

def bpnn(x,y,lr,epochs):
    print("Initial weights:")
    print(weights)
    for i in range(epochs):
        zh = weights["wh"]@x + weights["bh"]
        ah = sigmoid(zh)
        o = weights["wo"]@ah + weights["bo"]
        y_hat = sigmoid(o)
        loss = t_loss(y_hat,y)
        erroro = error(y,y_hat)
        delta = delt(y_hat)
        delto = erroro*delta
        dwo = delto@ah.T
        dbo = np.mean(delto)
        delth = weights["wo"].T@delto
        dwh = delth@x.T
        dbh = np.mean(delth,axis=1,keepdims=True)
        weights["wo"] = weights["wo"] - lr*(dwo/m)
        weights["bo"] = weights["bo"] - lr*(dbo/m)
        weights["wh"] = weights["wh"] - lr*(dwh/m)
        weights["bh"] = weights["bh"] - lr*(dbh/m)
        print(f"epoch:{i+1}, lr:{lr}, loss:{loss}")
    print("Updated weights:")
    print(weights)
```

```python
bpnn(x,y,lr = 0.001,epochs=1000)
```
